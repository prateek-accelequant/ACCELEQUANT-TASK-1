```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('real_data_1.csv')
# data used is: https://www.kaggle.com/datasets/ealtman2019/credit-card-transactions?resource=download
```

```python
df.head()
```

```python
df['Time'] = pd.to_timedelta(df['Time'].astype(str) + ':00')

# 2. Now your original math works perfectly!
df['time'] = df['Time'] / np.timedelta64(1, 'h')

# 3. Clean and convert the Amount column
df['Amount'] = df['Amount'].astype(str).str.replace('$', '', regex=False)
df['Amount'] = df['Amount'].str.replace(',', '', regex=False).str.strip()
df['amount'] = pd.to_numeric(df['Amount'], errors='coerce') 

# 4. Drop the original columns
df.drop(['Time', 'Amount'], axis=1, inplace=True)
```

```python
df.drop(['Merchant Name', 'Merchant City', 'Merchant State', 'Errors?'], axis=1, inplace=True)
```

```python
df = pd.get_dummies(df, columns=['Use Chip'], drop_first=True, dtype=int)
df['Is Fraud?'] = df['Is Fraud?'].map({'Yes': 1, 'No': 0})
```

```python
df.head(10)
```

```python
cols = ['Is Fraud?'] + [col for col in df.columns if col != 'Is Fraud?']
df = df[cols]
```

```python
df.dropna(inplace=True)
```

```python
df.to_csv('processed_fraud_data.csv', index=False)
```
