```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.circuit.library import ZZFeatureMap
import config
from models_quantum import ProductionQuantumKernelManager

print("=========================================================")
print("  QUANTUM CIRCUIT ARCHITECTURE VISUALIZATION HUB (QISKIT)  ")
print("=========================================================")

# ---------------------------------------------------------
print("\n[1] Rendering: Custom C and P Gates (Internal Structure)")
# ---------------------------------------------------------
c_gate_obj = ProductionQuantumKernelManager.get_custom_c_gate()
p_gate_obj = ProductionQuantumKernelManager.get_custom_p_gate()

print(" -> Custom C-Gate Internal Circuit Definition:")
display(c_gate_obj.definition.draw('mpl'))

print(" -> Custom P-Gate Internal Circuit Definition:")
display(p_gate_obj.definition.draw('mpl'))


# ---------------------------------------------------------
print("\n[2] Rendering: ZZ Feature Map (Linear vs Full Entanglement)")
# ---------------------------------------------------------
# ZZ Map with Linear Entanglement
zz_linear = ZZFeatureMap(
    feature_dimension=config.QUBIT_BUDGET, 
    reps=config.REPS, 
    entanglement='linear'
)
print(" -> ZZ Feature Map (Linear Entanglement - Decomposed Internal Gates):")
display(zz_linear.decompose().draw('mpl', fold=120))

# ZZ Map with Full Entanglement
zz_full = ZZFeatureMap(
    feature_dimension=config.QUBIT_BUDGET, 
    reps=config.REPS, 
    entanglement='full'
)
print(" -> ZZ Feature Map (Full Entanglement - Decomposed Internal Gates):")
display(zz_full.decompose().draw('mpl', fold=120))


# ---------------------------------------------------------
print("\n[3] Rendering: CPMap Feature Map (Black-Box C and P Gates)")
# ---------------------------------------------------------
mgr_cpmap_linear = ProductionQuantumKernelManager(map_type='CPMap')

print(" -> CPMap Feature Map (Linear Entanglement - C and P as Black Boxes):")
display(mgr_cpmap_linear.feature_map.draw('mpl', fold=120))

print("\n[Visualization Hub Complete] All circuit variations successfully rendered.")
```
